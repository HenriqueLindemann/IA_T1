```python
# Carregar o conjunto de dados CIFAR-10
cifar10 = keras.datasets.cifar10
#Carrega duas tuplas, representando os dados de treinamento e de teste.
#Cada tupla tem as imagens e os respectivos rótulos
(x_cifar10_train, y_cifar10_train), (x_cifar10_test, y_cifar10_test) = cifar10.load_data()
n_classes_cifar10 = 10

# Defina as classes do CIFAR-10
class_names = ['Avião', 'Automóvel', 'Pássaro', 'Gato', 'Cervo',
               'Cachorro', 'Sapo', 'Cavalo', 'Navio', 'Caminhão']

# Crie um dicionário para mapear as classes para as imagens correspondentes
class_to_image = {}
for i in range(10):
    index = (y_cifar10_test == i).nonzero()[0][0]  # Encontre o primeiro índice da classe
    class_to_image[i] = x_cifar10_test[index]

# Mostre uma imagem de cada classe
plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.xticks([])  # Remova os rótulos do eixo x
    plt.yticks([])  # Remova os rótulos do eixo y
    plt.imshow(class_to_image[i])
    plt.xlabel(class_names[i])

plt.tight_layout()
plt.show()

# Converter para codificação one-hot dos labels
#train_labels = tf.keras.utils.to_categorical(train_labels, num_classes=num_classes)
#test_labels = tf.keras.utils.to_categorical(test_labels, num_classes=num_classes)
#Não é necessário se utilizar como função de custo esparse_categorical_cross_entropy
```

```python
def get_fashion_mnist_network():
    network = tf.keras.models.Sequential([
        tf.keras.layers.Reshape((28, 28, 1), input_shape=(28, 28)),  # Reshape entrada para uma imagem 28x28x1
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same'),  # Convolutional layer
        tf.keras.layers.MaxPooling2D((2, 2)),  # Pooling layer
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Flatten(),  # Flatten layer
        tf.keras.layers.Dense(100, activation='relu'),  # Dense layer
        tf.keras.layers.Dropout(0.5),  # Dropout layer
        tf.keras.layers.Dense(n_classes_fmnist, activation='softmax')  # Output layer
    ])

    network.compile(optimizer='adam',
                    loss='sparse_categorical_crossentropy',
                    metrics=['accuracy'])

    network.summary()

    return network

def get_mnist_network():
    network = tf.keras.models.Sequential([
        tf.keras.layers.Reshape((28, 28, 1), input_shape=(28, 28)),  # Reshape entrada para uma imagem 28x28x1
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same'),  # Convolutional layer
        tf.keras.layers.MaxPooling2D((2, 2)),  # Pooling layer
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Flatten(),  # Flatten layer
        tf.keras.layers.Dense(100, activation='relu'),  # Dense layer
        tf.keras.layers.Dropout(0.5),  # Dropout layer
        tf.keras.layers.Dense(n_classes_mnist, activation='softmax')  # Output layer
    ])

    network.compile(optimizer='adam',
                    loss='sparse_categorical_crossentropy',
                    metrics=['accuracy'])

    network.summary()

    return network

def get_cifar100_network():
    model = keras.Sequential([
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)), # Convolutional layer
        tf.keras.layers.BatchNormalization(), # Batch layer para normalizacao de lote
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPooling2D((2, 2)), #Pooling layer
        tf.keras.layers.Dropout(0.2), # Dropout layer

        tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Dropout(0.3),

        tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Dropout(0.4),

        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(n_classes_cifar100, activation='softmax')  # 100 classes de saída
    ])

    # Compile o modelo
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    model.summary()

    return model

#Você pode aproveitar as ideias do exemplo acima
def get_cifar10_network():
    model = keras.Sequential([
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(32, 32, 3)), # Convolutional layer
        tf.keras.layers.BatchNormalization(), # Batch layer para normalizacao de lote
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPooling2D((2, 2)),  #Pooling layer
        tf.keras.layers.Dropout(0.2),  # Dropout layer

        tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Dropout(0.3),

        tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Dropout(0.4),

        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(n_classes_cifar10, activation='softmax')  # 10 classes de saída
    ])

    # Compile o modelo
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    model.summary()

    return model
```

```python
# Treine o modelo
model = get_cifar10_network()

start_time = time.time()
model.fit(x_cifar10_train, y_cifar10_train, epochs=10)

# Avalie o modelo no conjunto de teste
test_loss, test_accuracy = model.evaluate(x_cifar10_test, y_cifar10_test)
end_time = time.time()


time_taken_cifar10 = end_time - start_time
acc_cifar10 = test_accuracy * 100
print(f'Acurácia no conjunto de teste: {acc_cifar10:.2f}%')
print(f'Tempo de treinamento e avaliação: {time_taken_cifar10:.2f} segundos')
```

```python
# Treine o modelo
model = get_cifar100_network()

start_time = time.time()
model.fit(x_cifar100_train, y_cifar100_train, epochs=10)

# Avalie o modelo no conjunto de teste
test_loss, test_accuracy = model.evaluate(x_cifar100_test, y_cifar100_test)
end_time = time.time()


time_taken_cifar100 = end_time - start_time
acc_cifar100 = test_accuracy * 100
print(f'Acurácia no conjunto de teste: {acc_cifar100:.2f}%')
print(f'Tempo de treinamento e avaliação: {time_taken_cifar100:.2f} segundos')
```

```python
# Treine o modelo
model = get_mnist_network()

start_time = time.time()
model.fit(x_mnist_train, y_mnist_train, epochs=10)

# Avalie o modelo no conjunto de teste
test_loss, test_accuracy = model.evaluate(x_mnist_test, y_mnist_test)
end_time = time.time()

time_taken_mnist = end_time - start_time
acc_mnist = test_accuracy * 100
print(f'Acurácia no conjunto de teste: {acc_mnist:.2f}%')
print(f'Tempo de treinamento e avaliação: {time_taken_mnist:.2f} segundos')
```
